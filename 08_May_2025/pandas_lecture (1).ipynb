## Filtering berdasarkan kondisi

```python
df['City'].str.contains('City')

df[df['City'].str.contains('City')]
```

### Case filter kota yang density-nya diatas 15000

```python
# Boolean indexing
df[df['Density'] >= 15000]['City']
```

```python
# query method
df.query("Density >= 15000")['City']
```

### Case filter kota yang density-nya diatas 10000 dan area > 1000

```python
# boolean indexing
# untuk multiple condition bisa pakai dan (&) atau atau (|)
df[(df['Density'] >= 10000) & (df['Land_area'] > 1000)]
```

```python
# query method
# untuk multiple condition bisa pakai dan (and) atau atau (or)
df.query("Density > 10000 and Land_area >1000")
```

## Case tampilkan berapa banyak employee yang di hire tiap tahun

```python
# step 1 extract kolom tahun dari kolom hire_date

emp['year_hired'] = emp['hire_date'].dt.year
emp
```

```python
# step 2 grouping berdasarkan tahun
# menggunakan groupby()

emp.groupby('year_hired')['emp_id'].count().sort_values(ascending=False)
```

```python
emp.groupby('year_hired').agg({'emp_id': 'count',
                               'contoh': 'sum'})
```
