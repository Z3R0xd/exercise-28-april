```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd

# membuat object browser chrome
browser = webdriver.Chrome()

listNames = []
listYears = []
listWins = []

#1-10
for number in range(1,25):
    # membuka website
    browser.get("https://www.scrapethissite.com/pages/forms/?page_num=" + str(number))

    # extract HTML satu halaman
    html = browser.page_source

    # parsing string ke BeautifulSoup object
    page = BeautifulSoup(html, "html.parser")

    # method untuk mencari element
    # find(nama_element, attribute) -> return 1 element yang pertama kali ditemukan
    # find_all(nama_element, attribute) -> return >1 element yang ditemukan

    # <td class="name"> Boston Bruins </td>
    # find("td")
    # find("td", {"class": "name"})

    result = page.find_all("tr", {"class":"team"})
    # print(result)

    for el in result:
        # extract content from every element
        # get_text() -> return content from element
        # content = el.get_text().strip()
        # listYears.append(content)
        
        # extract name element
        nameEl = el.find("td", {"class":"name"}) # -> return element 
        # extract year element
        yearEl = el.find("td",{"class":"year"})
        # extract win element
        winEl = el.find("td",{"class":"wins"})
        
        # extract content setiap element
        # validasi element exists
        if nameEl != None:
            # extract content dan menambahkan ke list
            listNames.append(nameEl.get_text().strip())    
        else:
            # menambahkan None ke list
            listNames.append(None)
        
        if yearEl != None:
            # extract content dan menambahkan ke list
            listYears.append(yearEl.get_text().strip())    
        else:
            # menambahkan None ke list
            listYears.append(None)
            
        if winEl != None:
            # extract content dan menambahkan ke list
            listWins.append(winEl.get_text().strip())    
        else:
            # menambahkan None ke list
            listWins.append(None)
    
# bikin dataframe
df = pd.DataFrame()
# input data
df["Years"] = listYears
df["Names"] = listNames
df["Wins"] = listWins

display(df)
# menutup browser
browser.quit()
```
